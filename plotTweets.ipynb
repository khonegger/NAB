# Anomaly Detection: Milestone #3

Visualizations of Numenta NAB anomaly detection benchmark dataset. This particular benchmark subset contains ~2 months of tweet volume data for 10 stock symbols at 5 min intervals.

The scripts below load the data and plot the time series, with labeled anomalies overlayed in red.

```python
import scipy as scipy
import numpy as np
import pandas as pd
import json
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Load JSON file containing anomaly labels for all data
with open('labels/combined_labels.json') as jf:
    anomLabels = json.load(jf)
```

```python
# Function: plots time series for single stock symbol, takes csv file name as input

def plotTwitterTimeSeries(data_fname):
    
    base_path = 'data/realTweets/'
    
    
    # Read NAB data into pandas dataframe
    tweets = pd.read_csv(base_path + data_fname)
    
    
    # Match anomaly timestamps to timeseries indices
    tweet_vals = tweets['value']
    anom_tstamp = anomLabels['realTweets/' + data_fname]

    anom_idx  = [tweet_vals[tweets['timestamp'] == x].index.values[0] for x in anom_tstamp]
    anom_vals = [tweet_vals[tweets['timestamp'] == x].as_matrix()[0] for x in anom_tstamp]
    
    
    # Plot tweet volume time series, highlight anomalies
    plt.figure(figsize=[20,10])

    plt.plot(tweets['value'], zorder=1)
    plt.hold
    plt.scatter(anom_idx, anom_vals, s=150, c='red', zorder=2)
    plt.xlim([0,tweets.shape[0]])
    plt.title(data_fname[:-4])
    plt.xlabel('sample index')
    plt.ylabel('tweet volume')
```

```python
data_fname = 'Twitter_volume_AAPL.csv'
plotTwitterTimeSeries(data_fname)
```
